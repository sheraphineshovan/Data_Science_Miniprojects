```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
train=pd.read_csv('/content/fashion-mnist_train.csv')
test=pd.read_csv('/content/fashion-mnist_test.csv')
```

```python
train.shape
```

```python
test.shape
```

```python
train.info(max_cols=785)
```

```python
x=train.iloc[:,1:]
y=train["label"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=39)
```

```python
def report(x,y):
  print(accuracy_score(x,y))
  print(confusion_matrix(x,y))
  print(classification_report(x,y))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=350,min_samples_split=5,verbose=1)
rfc.fit(x_train,y_train)
```

```python
ypred_train_rfc=rfc.predict(x_train)
ypred_test_rfc=rfc.predict(x_test)
report(y_train,ypred_train_rfc)
report(y_test,ypred_test_rfc)
```

```python
slp=Perceptron(verbose=1)
slp.fit(x_train,y_train)
```

```python
ypred_slp_train=slp.predict(x_train)
ypred_slp_test=slp.predict(x_test)
```

```python
report(y_train,ypred_slp_train)
report(y_test,ypred_slp_test)
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(64,64,64),verbose=1,max_iter=250)
mlp.fit(x_train,y_train)
```

```python
ypred_train_mlp=mlp.predict(x_train)
ypred_test_mlp=mlp.predict(x_test)
```

```python
report(y_train,ypred_train_mlp)
report(y_test,ypred_test_mlp)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.optimizers import SGD,Adam
from keras.losses import categorical_crossentropy
from tensorflow.keras.utils import to_categorical
```

```python
model=Sequential()
```

```python
model.add(Dense(512,activation='relu',input_dim=784))
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
y_train_ohe=to_categorical(y_train)
y_test_ohe=to_categorical(y_test)
```

```python
y_train_ohe.shape
```

```python
y_test_ohe.shape
```

```python
  hist=model.fit(x_train,y_train_ohe,validation_data=(x_test,y_test_ohe),epochs=200,batch_size=64,verbose=2)
```
